```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv2.imread("1.jpg")
background = cv2.imread("2.jpg")
plt.imshow(image)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

## Blurring the whole image

```python
a = np.copy(image)
blur_image = cv2.blur(a, (10, 10))
plt.imshow(blur_image)
```

```python
greenl = np.array([0,50,0])
greenu = np.array([100,255,100])
mask = cv2.inRange(a, greenl, greenu)
plt.imshow(mask,'gray')
```

```python
lower_blue = np.array([0, 200, 0])    
upper_blue = np.array([220, 255, 200])
mask1 = cv2.inRange(a, lower_blue,upper_blue)
plt.imshow(mask1,'gray')
```

## Converting the background to white

```python
masked_image = np.copy(a)
masked_image[mask != 0] = [255,255, 255]

plt.imshow(masked_image)
```

## Converting the background to RED


```python
masked_image = np.copy(a)
masked_image[mask != 0] = [255,0, 0]

plt.imshow(masked_image)
```
