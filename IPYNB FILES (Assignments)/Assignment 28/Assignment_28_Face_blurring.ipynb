```python
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt
 

image = cv2.cvtColor(cv2.imread('/content/ADARSH.jpeg') , cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
detector = MTCNN()
faces = detector.detect_faces(image)
for face in faces:
  bounding_box = faces[0]['box']
  x = bounding_box[0]
  y = bounding_box[1]
  x1 = bounding_box[2]
  y1 = bounding_box[3]



  rectangle = cv2.rectangle(image,
              (x,y),
              (x+x1,y+y1),
              (0,0,0),2)

  x,y = y,x
  x1,y1 = y1,x1
  blur = rectangle[x:x+x1,y:y+y1]
  image[y:y+blur.shape[0], x:x+blur.shape[1]] = blur

  x,y = y,x
  x1,y1 = y1,x1


  blur_image = cv2.blur(blur, (1000, 1000))

  rectangle[y:y+blur.shape[0], x:x+blur.shape[1]] = blur_image
  blur_image = rectangle[y:y+blur.shape[0], x:x+blur.shape[1]]
  




  # plt.imshow(blur_image)
plt.imshow(image)
```
