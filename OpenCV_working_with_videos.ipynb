```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
video = cv2.VideoCapture("stock_video.mp4")
size = (int(video.get(3)), int(video.get(4)))

fourcc = cv2.VideoWriter_fourcc(*'VP90')

out = cv2.VideoWriter(
    "output.webm", fourcc, 25.0, size, 0)


def process_frame(frame):
    return cv2.Canny(frame, 20, 100)


while True:
    ret, frame = video.read()

    if not ret:
        break

    frame = process_frame(frame)

    plt.imshow(frame, cmap='gray')
    plt.show()

    break

    out.write(frame)


video.release()
out.release()
```
