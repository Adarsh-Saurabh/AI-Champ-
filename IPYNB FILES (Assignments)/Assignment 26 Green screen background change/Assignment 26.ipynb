```python
import numpy
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('2.jpg')
plt.imshow(image)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
window_name = 'Image'
font = cv2.FONT_HERSHEY_SIMPLEX
org = (50, 50)
fontScale = 1
color = (255, 155, 191)
thickness = 2
image = cv2.putText(image, 'AtomBomb', org, font, 
                   fontScale, color, thickness, cv2.LINE_AA)
   
# Displaying the image
plt.imshow(image) 
```
