```python
model = torch.hub.load('feat7/yolov5', 'custom', path_or_model='..write path here..last.pt')
model.autoshape()
```

```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.cvtColor(cv2.imread("/content/train/images/Smog-Mask-PM2-5-Carbon-Filter-Face-Masks-Fog-Dust-Air-Cleaner-anti-pollution-Face-Protector_jpg.rf.8b5dc4859890d34642d44ee9d1e7a1e8.jpg"), cv2.COLOR_BGR2RGB)
res = model(image)
```

```python
res.xywh
```

```python
draw_rect = image.copy()

for result in res.xyxy[0].tolist():
  x1, y1, x2, y2, confidence, class_name = result

  x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

  print(result)
  draw_rect = cv2.rectangle(draw_rect, (x1, y1), (x2, y2), (255, 0, 0), 1)

plt.imshow(draw_rect)
```
