```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

# What is an image?

### Read and see images

```python
image = cv2.imread("gift_box.png")
plt.imshow(image)
```

### Color images and Grayscale images

```python
color_image = cv2.imread("gift_box.png")
color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)
plt.imshow(color_image)
```

```python
grayscale_image = cv2.imread("gift_box.png", 0)
plt.imshow(grayscale_image, cmap='gray')
```

```python
## Color image has 3 channels: R, G and B while grayscale image has only one channel

print(f"Color image shape: {color_image.shape}\nGrayscale image shape: {grayscale_image.shape}")
```

### See individual channels (R, G, B)

```python
print("Channel 1 (Blue): ")
plt.imshow(color_image[:, :, 0], cmap='gray')
plt.show()

print("Channel 2 (Green): ")
plt.imshow(color_image[:, :, 1], cmap='gray')
plt.show()

print("Channel 3 (Red): ")
plt.imshow(color_image[:, :, 2], cmap='gray')
plt.show()
```

# Understanding Image Coordinate System

```python
color_image = cv2.imread("gift_box.png")
color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)
plt.imshow(color_image)
```

# Playing with image pixels

```python
a = np.copy(color_image)
plt.imshow(a)
```

```python
a[300:700, 180:220] = 0
plt.imshow(a)
```

```python
lower_white = np.array([220, 220, 220], dtype=np.uint8)
upper_white = np.array([255, 255, 255], dtype=np.uint8)
mask = cv2.inRange(a, lower_white, upper_white)

a[mask == 255] = (125, 255, 120)
plt.imshow(a)
```

# Convolution

[Digarams to understand cv basics](https://drive.google.com/file/d/1S05oKga8AGwak12-XMknAD6MqGaGEJ8K/view?usp=sharing)

```python
a = np.copy(color_image)

kernel = np.array([
                   [-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1],

])

b = cv2.filter2D(a, -1, kernel)
plt.imshow(b)
```

```python
a = np.copy(color_image)

kernel = np.array([
                   [-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1],

])

b = cv2.filter2D(a, -1, kernel)
plt.imshow(b)
```

```python
a = np.copy(color_image)

kernel = np.array([
	[0, 1, 0],
	[1, -4, 1],
	[0, 1, 0]
])

b = cv2.filter2D(a, -1, kernel)
plt.imshow(b)
```

# Canny edge detection

```python
a = np.copy(color_image)

b = cv2.Canny(a, 50, 100)

plt.imshow(b, cmap='gray')
```

# Blurring

```python
a = np.copy(color_image)
blur_image = cv2.blur(a, (100, 100))
plt.imshow(blur_image)
```
